```python

# --- Imports ---
import os
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OrdinalEncoder, StandardScaler
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import (roc_auc_score, roc_curve, classification_report,
                             confusion_matrix, accuracy_score)

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import (Input, Dense, Dropout, BatchNormalization,
                                     Activation, Bidirectional, LSTM, LayerNormalization,
                                     SeparableConv1D)
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.regularizers import l2
from tensorflow.keras.metrics import AUC
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint

# --- Settings ---
sns.set(style="whitegrid")
import warnings
warnings.filterwarnings("ignore")

```

```python
# =========================================================
# Reproducibility
# =========================================================
SEED = 42
random.seed(SEED)
np.random.seed(SEED)
tf.random.set_seed(SEED)
os.environ['PYTHONHASHSEED'] = str(SEED)
```

```python
# =========================================================
# Parameters & Paths
# =========================================================
DATA_PATH = "Disorders_and_Diagnosis_EEG_Dataset_v.csv"
N_SPLITS = 5
EPOCHS = 100
BATCH_SIZE = 16
CLIP_NORM = 1.0
MODEL_DIR = "./models_lstm_cv"
os.makedirs(MODEL_DIR, exist_ok=True)

print("‚úÖ Setup complete. Ready to train model.")
```

```python
# =========================================================
# Helper: Column name reformatter
# =========================================================
def reformat_name(name):
    """Simplify long EEG feature names into shorter format"""
    splitted = name.split(".")
    if len(splitted) >= 5 and splitted[0] != "COH":
        return f"{splitted[2]}.{splitted[4]}"
    elif len(splitted) >= 7 and splitted[0] == "COH":
        return f"{splitted[0]}.{splitted[2]}.{splitted[4]}.{splitted[6]}"
    return name
```

```python
# =========================================================
# Load & Preprocess Data
# =========================================================
df = pd.read_csv(DATA_PATH)
df.rename(reformat_name, axis=1, inplace=True)

# Select only Anxiety disorder vs Healthy control
df = df[df["main.disorder"].isin(["Anxiety disorder","Healthy control"])].reset_index(drop=True)

# EEG Features of interest (FP1, FP2 + coherence)
fp_features = [
    "delta.FP1","delta.FP2","theta.FP1","theta.FP2",
    "alpha.FP1","alpha.FP2","beta.FP1","beta.FP2",
    "highbeta.FP1","highbeta.FP2","gamma.FP1","gamma.FP2"
]
coh_features = [
    "COH.delta.FP1.FP2","COH.theta.FP1.FP2","COH.alpha.FP1.FP2",
    "COH.beta.FP1.FP2","COH.highbeta.FP1.FP2","COH.gamma.FP1.FP2"
]
selected_features = [c for c in fp_features+coh_features if c in df.columns]

# Metadata
meta_cols = [c for c in ["sex","education","IQ","age"] if c in df.columns]
use_cols = ["main.disorder","specific.disorder"] + meta_cols + selected_features
df = df[use_cols].copy()

print("‚úÖ Dataset loaded with shape:", df.shape)
```

```python
# =========================================================
# Handle Missing Values
# =========================================================
imputer = SimpleImputer(strategy="median")
mv_cols = [c for c in ["education","IQ"] if c in df.columns]
if mv_cols:
    df[mv_cols] = imputer.fit_transform(df[mv_cols])

# Labels
df["label"] = df["main.disorder"].map({"Healthy control":0,"Anxiety disorder":1})

# Encode categorical variables
feature_meta_cols = []
if "sex" in df.columns:
    enc = OrdinalEncoder()
    df["sex_enc"] = enc.fit_transform(df[["sex"]])
    feature_meta_cols.append("sex_enc")

if "age" in df.columns:
    df["age"] = pd.to_numeric(df["age"], errors="coerce")
    df["age"].fillna(df["age"].median(), inplace=True)
    feature_meta_cols.append("age")

# Ensure numeric EEG features
for c in selected_features:
    df[c] = pd.to_numeric(df[c], errors="coerce")
    df[c].fillna(df[c].median(), inplace=True)

# Log-transform
df[selected_features] = np.log1p(df[selected_features].astype(float))

# Final feature list
features = feature_meta_cols + selected_features
X_all = df[features].values
y_all = df["label"].values

# Standard scaling
scaler = StandardScaler()
X_all = scaler.fit_transform(X_all)

# Reshape for LSTM
X_all = X_all.reshape((X_all.shape[0], X_all.shape[1], 1))
print("‚úÖ Final dataset shape (samples, timesteps, channels):", X_all.shape)
```

```python
import joblib
joblib.dump(scaler, "scaler.pkl")
```

```python
# =========================================================
# Dataset Details & Exploration
# =========================================================

print("üîé Dataset Overview")

# Shape
print(f"Number of samples: {df.shape[0]}")
print(f"Number of columns: {df.shape[1]}")

# Columns
print("\nüìã Columns in dataset:")
print(df.columns.tolist())

# Target distribution
print("\nüéØ Target label distribution:")
print(df["main.disorder"].value_counts())

# Check missing values
print("\n‚ö†Ô∏è Missing values per column:")
print(df.isnull().sum())

# Show first few rows
print("\nüîπ First 5 rows of dataset:")
display(df.head())

# =========================================================
# Visualizations of Dataset
# =========================================================

# 1. Class distribution (Healthy vs Anxiety)
plt.figure(figsize=(6,4))
sns.countplot(data=df, x="main.disorder", palette="Set2")
plt.title("Class Distribution")
plt.xlabel("Disorder")
plt.ylabel("Count")
plt.show()

# 2. Age distribution (if available)
if "age" in df.columns:
    plt.figure(figsize=(6,4))
    sns.histplot(df["age"], kde=True, bins=20, color="teal")
    plt.title("Age Distribution")
    plt.xlabel("Age")
    plt.ylabel("Count")
    plt.show()

# 3. Sex distribution (if available)
if "sex" in df.columns:
    plt.figure(figsize=(5,4))
    sns.countplot(data=df, x="sex", palette="Set1")
    plt.title("Sex Distribution")
    plt.xlabel("Sex")
    plt.ylabel("Count")
    plt.show()

# 4. Feature correlation heatmap (optional: only EEG features)
plt.figure(figsize=(10,8))
sns.heatmap(df[selected_features].corr(), cmap="coolwarm", center=0)
plt.title("Correlation Heatmap of EEG Features")
plt.show()
```

```python
# =========================================================
# Model Builder
# =========================================================
def build_lstm_model(input_timesteps, input_features, seed=SEED):
    """Builds and compiles a lightweight BiLSTM model"""
    tf.keras.backend.clear_session()
    tf.random.set_seed(seed)
    
    model = Sequential([
        Input(shape=(input_timesteps, input_features)),

        SeparableConv1D(
            16, kernel_size=3, padding="same",
            depthwise_regularizer=l2(1e-4),
            pointwise_regularizer=l2(1e-4)
        ),
        BatchNormalization(),
        Activation("relu"),
        Dropout(0.1),

        Bidirectional(
            LSTM(32, return_sequences=False, recurrent_dropout=0.1,
                 kernel_regularizer=l2(1e-4))
        ),
        LayerNormalization(),
        Dropout(0.25),

        Dense(16, activation="relu", kernel_regularizer=l2(1e-4)),
        Dropout(0.2),

        Dense(1, activation="sigmoid")
    ])

    opt = Adam(learning_rate=1e-3, clipnorm=CLIP_NORM)
    model.compile(
        optimizer=opt,
        loss="binary_crossentropy",
        metrics=["accuracy", AUC(name="auc")]
    )
    return model
```

```python
# =========================================================
# Cross-validation Training
# =========================================================
skf = StratifiedKFold(n_splits=N_SPLITS, shuffle=True, random_state=SEED)

oof_probs = np.zeros(len(X_all))
oof_preds = np.zeros(len(X_all))
fold_histories = []
fold_metrics = []

for fold, (train_idx, val_idx) in enumerate(skf.split(X_all, y_all), 1):
    print(f"\nüîπ Training Fold {fold}/{N_SPLITS} üîπ")

    X_tr, X_val = X_all[train_idx], X_all[val_idx]
    y_tr, y_val = y_all[train_idx], y_all[val_idx]

    # Class weights
    classes, counts = np.unique(y_tr, return_counts=True)
    class_weight = {int(c): float(len(y_tr) / (len(classes) * cnt)) for c, cnt in zip(classes, counts)}
    print("Class weights:", class_weight)

    # Build model
    model = build_lstm_model(X_all.shape[1], X_all.shape[2], seed=SEED+fold)

    # Callbacks
    fold_model_path = os.path.join(MODEL_DIR, f"lstm_fold{fold}_best.keras")
    es = EarlyStopping(monitor="val_auc", mode="max", patience=12, restore_best_weights=True, verbose=1)
    rlrop = ReduceLROnPlateau(monitor="val_auc", mode="max", factor=0.5, patience=6, min_lr=1e-6, verbose=1)
    mc = ModelCheckpoint(fold_model_path, monitor="val_auc", mode="max", save_best_only=True, verbose=1)

    # Train
    history = model.fit(
        X_tr, y_tr,
        validation_data=(X_val, y_val),
        epochs=EPOCHS,
        batch_size=BATCH_SIZE,
        class_weight=class_weight,
        callbacks=[es, rlrop, mc],
        verbose=2
    )

    # Save history
    fold_histories.append(history.history)

    # Predict validation
    val_probs = model.predict(X_val).ravel()
    val_preds = (val_probs > 0.5).astype(int)

    oof_probs[val_idx] = val_probs
    oof_preds[val_idx] = val_preds

    # Metrics
    auc = roc_auc_score(y_val, val_probs)
    acc = accuracy_score(y_val, val_preds)
    print(f"‚úÖ Fold {fold} Results: AUC={auc:.4f}, ACC={acc:.4f}")
    fold_metrics.append({"fold": fold, "auc": auc, "acc": acc})
```

```python
# =========================================================
# Final Evaluation
# =========================================================
print("\nüìä Cross-validation Results")
df_metrics = pd.DataFrame(fold_metrics)
print(df_metrics)
print("Mean AUC:", df_metrics["auc"].mean())
print("Mean ACC:", df_metrics["acc"].mean())

# Classification report on OOF
print("\nüîé Overall Classification Report")
print(classification_report(y_all, oof_preds))

# Confusion Matrix
cm = confusion_matrix(y_all, oof_preds)
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
plt.title("Confusion Matrix (OOF Predictions)")
plt.xlabel("Predicted")
plt.ylabel("True")
plt.show()
```

```python
# =========================================================
# Visualizations
# =========================================================

# --- 1. Training curves for each fold ---
# Accuracy
plt.figure(figsize=(14,6))
for i, h in enumerate(fold_histories, 1):
    plt.plot(h["accuracy"], label=f"Fold {i} Train Acc")
    plt.plot(h["val_accuracy"], linestyle="--", label=f"Fold {i} Val Acc")
plt.title("Training vs Validation Accuracy per Fold")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend()
plt.show()

# AUC
plt.figure(figsize=(14,6))
for i, h in enumerate(fold_histories, 1):
    plt.plot(h["auc"], label=f"Fold {i} Train AUC")
    plt.plot(h["val_auc"], linestyle="--", label=f"Fold {i} Val AUC")
plt.title("Training vs Validation AUC per Fold")
plt.xlabel("Epoch")
plt.ylabel("AUC")
plt.legend()
plt.show()

# Loss
plt.figure(figsize=(14,6))
for i, h in enumerate(fold_histories, 1):
    plt.plot(h["loss"], label=f"Fold {i} Train Loss")
    plt.plot(h["val_loss"], linestyle="--", label=f"Fold {i} Val Loss")
plt.title("Training vs Validation Loss per Fold")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.show()

# --- 2. ROC Curve (OOF predictions) ---
fpr, tpr, _ = roc_curve(y_all, oof_probs)
plt.figure(figsize=(6,6))
plt.plot(fpr, tpr, label=f"AUC = {roc_auc_score(y_all, oof_probs):.3f}")
plt.plot([0,1], [0,1], "--", color="gray")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("Overall ROC Curve (OOF Predictions)")
plt.legend()
plt.show()

# --- 3. Precision, Recall, F1 ---
from sklearn.metrics import precision_score, recall_score, f1_score

precision = precision_score(y_all, oof_preds)
recall = recall_score(y_all, oof_preds)
f1 = f1_score(y_all, oof_preds)

print("\nüìä Final Evaluation Metrics (OOF Predictions)")
print(f"Accuracy : {accuracy_score(y_all, oof_preds):.4f}")
print(f"AUC      : {roc_auc_score(y_all, oof_probs):.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall   : {recall:.4f}")
print(f"F1-score : {f1:.4f}")
```
